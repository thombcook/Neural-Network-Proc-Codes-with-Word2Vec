```python
#Testing the creation of this model for members.

from gensim.models.word2vec import Word2Vec


num_features = 8  # Word vector dimensionality
min_word_count = 1 # Minimum word count
num_workers = 6     # Number of parallel threads
context = 3        # Context window size
epochs = 6          # number of epochs 
sg = 0               # skipgram model (if 0, then CBOW)    
model_embedder = Word2Vec(master_proc_list, min_count=min_word_count, size = num_features,workers=num_workers,
                 iter=epochs,window =context, sg= sg) 
```

```python
proc_list_stuff = []
for i in master_proc_list:
    for j in i:
        if j not in proc_list_stuff:
            proc_list_stuff.append(j)
```

```python
#Make a list of the words in order of frequency
word_list_in_order = []
for i in range(len(model_embedder.wv.vocab)):
    word_list_in_order.append(model_embedder.wv.index2word[i])
```

```python
#Fill an array with the vectors. Doing this because can't tell what the column names are for the inbuilt gensim function. 
template = np.zeros(model_embedder.wv.vectors.shape)

for i, j in enumerate(word_list_in_order):
    template[i] = model_embedder[j]
#Make this into a DataFrame!
templatedf = pd.DataFrame(template.T, columns = word_list_in_order)
```
